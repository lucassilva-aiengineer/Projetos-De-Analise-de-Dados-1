```python
import numpy as np
```

```python
qntd_infracoes_tipo = np.array([grupos.MEDIA,
                                grupos.GRAVE,
                                grupos.GRAVISSIMA,
                                grupos.LEVE])


tipos = ['MÃ©dia', 'Grave', 'Gravissima', 'Leve']
```

```python

cores = ["#ffff10",
         "#ff6518",
         "#ff1212",
         "#8fff18"]


plt.pie(qntd_infracoes_tipo,
        labels= tipos,
        autopct= "%1.f%%",
        colors= cores,
        explode= [0, 0, 0.2, 0],
        shadow= True,
        startangle= 190
        )

plt.title("GOIÃS - TIPO DAS INFRAÃ‡Ã•ES COMETIDAS ",
          fontsize= 20,
          family= 'sans-serif',
          fontweight= 'bold',
          color= 'darkorange'
          )


plt.show()

# Storytelling com dados.
```

```python
# As cinco cidades de GoiÃ¡s com as maiores quantidades de infraÃ§Ãµes:
quantidades = np.array(list(serie_localidade_contagem.iloc[:5]))


# quantidades_a = [13492, 8778, 7904,
#                6661, 5357]

# quantidades_a =

valores_a =  ['GoiÃ¢nia',
            'Ap GoiÃ¢nia',
            'BrasÃ­lia',
            'AnÃ¡polis',
            'Rio Verde']

plt.bar(valores_a, quantidades,
        color= 'darkorange')

plt.title("Cidades Com mais casos de multas",
          fontsize= 20,
          family= 'sans-serif',
          color= 'Orange',
          fontweight= 'bold')

plt.xlabel('Cidades',
           fontsize= 15,
           family= 'sans-serif',
           color= 'orange',
           fontweight= 'bold')

family_a = 'sans-serif'
plt.ylabel('NÃºmero Multas',
           fontsize= 15,
           color= 'orange',
           family= family_a,
           fontweight= 'bold')


plt.show()
```

```python
# combustiveis_fosseis = df[df['combustivel'].isin[else for string_combustivel in combustiveis_tipos]]

from typing import Dict

# Eu vou separar os tipos de combustiveis e depois plotar um histograma, verificando como está a distribuição
# dos tipos de combustíveis na frota paulista de automóveis.

def classificar():
    """Um função que classifica os tipos de combustíveis
      em três classificações: Hibridos, Elétricos e Fósseis. """

    dicionario_classificacao_tipos: Dict[str, list] = {}

    for string_combustivel in lista_combustiveis:


        if "eletrico" in string_combustivel.lower():

            if not "combustiveis_renovaveis" in dicionario_classificacao_tipos:
                dicionario_classificacao_tipos["combustiveis_renovaveis"] = []
                dicionario_classificacao_tipos["combustiveis_renovaveis"].append(string_combustivel)

            else:
                dicionario_classificacao_tipos["combustiveis_renovaveis"].append(string_combustivel)

        elif "gas" in string_combustivel.lower():

            if not "combustiveis_fosseis" in dicionario_classificacao_tipos:
                dicionario_classificacao_tipos["combustiveis_fosseis"] = []
                dicionario_classificacao_tipos["combustiveis_fosseis"].append(string_combustivel)

            else:
                dicionario_classificacao_tipos["combustiveis_fosseis"].append(string_combustivel)

        elif "diesel" in string_combustivel.lower():

            if not "combustiveis_fosseis" in dicionario_classificacao_tipos:

                dicionario_classificacao_tipos["combustiveis_fosseis"] = []
                dicionario_classificacao_tipos["combustiveis_fosseis"].append(string_combustivel)

            else:
                dicionario_classificacao_tipos["combustiveis_fosseis"].append(string_combustivel)

        elif "alcool" in string_combustivel.lower():

            if not "combustiveis_renovaveis" in dicionario_classificacao_tipos:

                dicionario_classificacao_tipos["combustiveis_renovaveis"] = []
                dicionario_classificacao_tipos["combustiveis_renovaveis"].append(string_combustivel)

            else:
                dicionario_classificacao_tipos["combustiveis_renovaveis"].append(string_combustivel)


    return dicionario_classificacao_tipos

```

```python
print(classificar())
```

```python
dicionario_classificacao = classificar()

dicionario_classificacao["combustiveis_renovaveis"]
```

```python
# Filtrando os veículos movidos a energia renovável ou não.

df_renovavel = df[df['combustivel'].isin(dicionario_classificacao["combustiveis_renovaveis"])]
df_fossel = df[df['combustivel'].isin(dicionario_classificacao["combustiveis_fosseis"])]
```

```python
from typing import List
```

```python
# Criando um histograma comparando a distribuição dos tipos de combustíveis.

combustiveis = ["Renováveis", "Não Renováveis"]

quantidade_autm_renovaveis = len(df_renovavel)
quantidade_autm_df_fossel = len(df_fossel)

# quantidades_veiculos_combustivel = [quantidade_autm_renovaveis, quantidade_autm_df_fossel]

qntd_veiculos_combustivel: List[int] = []
qntd_veiculos_combustivel.append(quantidade_autm_renovaveis)

# qntd_veiculos_combustivel.extend([quantidade_autm_df_fossel])
qntd_veiculos_combustivel += [quantidade_autm_df_fossel]

plt.bar(range(len(combustiveis)), qntd_veiculos_combustivel
                                ,color= '#2da47d')

plt.title("Veículos Combustíveis - SP"
                                          ,color= '#2da47d'
                                          ,family= 'sans'
                                          ,fontsize= 20
                                          ,fontweight= 'bold')

plt.ylabel("Número Veículos"
           ,color= '#2da47d')

plt.tick_params(axis= "both",
                colors= "#2da47d")


plt.axis([-0.5, 1.5, 0, 100000])

plt.xticks(range(len(combustiveis)), combustiveis)

plt.grid(True, linestyle='--', alpha= 0.9)

plt.show()
```
